```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

llm = ChatGroq(
    model="llama-3.1-8b-instant"        
)

response = llm.invoke("Did you know Andhra")
print(response.content) 
```

# Message Prompts

Message Types:
1. System message - Tells models how to behave
2. Human message - User input and interactions
3. AI message - Response gen by AI 
4. Tool message - Output of tools 


ROLE
CONTENT

```python
from langchain.messages import SystemMessage, HumanMessage, ToolMessage, AIMessage

messages=[
    SystemMessage("You are a poetry expert"),
    HumanMessage("What is AWS")
]

response=llm.invoke(messages)
response.content
```

```python
system_msg = SystemMessage("You are a helpful coding assistant") 

messages = [
    system_msg,
    HumanMessage("What is REST API?")
]

response=llm.invoke(messages)
print(response.content)
response.usage_metadata
```

With Big System message

```python
system_msg = SystemMessage("Here long prompt for what you need ---------")  #More prompt more accuracy results

messages = [
    system_msg,
    HumanMessage("What is REST API?")
]

response=llm.invoke(messages)
print(response.content)
```

```python
from langchain.messages import SystemMessage, HumanMessage, ToolMessage, AIMessage

messages=[
    SystemMessage("You are a poetry expert"),
    HumanMessage("What is AWS")
]

response=llm.invoke(messages)
response.content
```

```python
from langchain.messages import SystemMessage, HumanMessage, ToolMessage, AIMessage

ai_msg = AIMessage(
    content=[],
    tool_calls=[{
        "name": "get-weather",
        "args": {"location:": "India, Andhrapradesh"},
        "id": "call_123"
    }]
)

weather_result = "Sunny", "44 C"
tool_message= ToolMessage(
    content=weather_result,
    tool_call_id="call_123"
)

messages=[
    
    HumanMessage("What is AWS"),
    ai_msg,
    tool_message,
]

response=llm.invoke(messages)
response.content
```
