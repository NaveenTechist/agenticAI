Tools


Model can be request to call tools that perform tasks like fetching data from a database, searching the web, running code.  

```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

llm = ChatGroq(
    model="llama-3.1-8b-instant"    
)

response = llm.invoke("Did you know Andhra")
print(response.content)
```

```python
from langchain.tools import tool

@tool 
def get_weather(location: str) -> str:
    """Get the weather at a location"""
    return f"It's sunny in {location}"

model_with_tools = llm.bind_tools([get_weather])

response=model_with_tools.invoke("What is weather like in Boston?") 
print(response)
for i in response.tool_calls:
    # View tool calls made by model
    print(f"Tool: {tool_calls['name']}")
    print(f"Args: {tool_calls['args']}")
```

# Tool Execution Loops

```python
# Step - 1 Model generates tools

messages = [{"role":"user", "content": "What is the weather in Boston?"}] 
ai_msg = model_with_tools.invoke(messages)
messages.append(ai_msg)  

#Step - 2: Execute tools and collect results 
for tool in ai_msg.tool_calls: 
    tool_result = get_weather.invoke(tool)
    messages.append(tool_result)

# Step-3: Pass results back to model for final response 
final_response = model_with_tools.invoke(messages) 
print(final_response.text)   
```
