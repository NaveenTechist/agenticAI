# Structured Output 


Models can be requested to provide their responses in a format matching given schema. This is usefull for enusuring the output  can easily parsed and used for processing.

Techniques:
1. Pydantic
2. TypedDict

# Pydantic

Pydantic models provide the richest feature set with field validation, desciption and nested structures.


```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq

load_dotenv()

llm = ChatGroq(
    model="llama-3.1-8b-instant"        
)
response=llm.invoke("Hello")
print(response)
```

```python
from pydantic import BaseModel, Field 

class Movie(BaseModel):
    title:str=Field(description="The title of the movie")
    year:str=Field(description="This is year of the movie")
    director:str=Field(description="The director of the movie")
    rating:str=Field(description="The movie rating of out of 10")
```

```python
model_with_structured=llm.with_structured_output(Movie)
model_with_structured
```

```python
respnse=model_with_structured.invoke("Provide details about the movie Inception")
respnse
```

```python
from pydantic import BaseModel, Field 

class Movie(BaseModel):
    title:str=Field(description="The title of the movie")
    year:str=Field(description="This is year of the movie")
    director:str=Field(description="The director of the movie")
    rating:str=Field(description="The movie rating of out of 10")

model_with_structured = llm.with_structured_output(Movie, include_raw=True)

response=model_with_structured.invoke("Provide details about the movie Inception")
response
```

# TypedDict
